```python
def get_weighted_price(row, df_april):
    # Название проекта
    project = row['ЖК рус']
    section = row['Корпус']
    area = row['Площадь']

    # Основной фильтр: по названию проекта
    match = df_april[df_april['Название проекта'] == project.str.lower()]

    # Дополнительный фильтр по корпусу, если секция указана
    if pd.notna(section):
        match = match[match['Корпус'] == section]

    # Поиск точного совпадения по площади
    exact = match[np.isclose(match['Площадь, кв.м'], area, atol=0.01)]

    if not exact.empty:
        # Возвращаем первую найденную цену при точном совпадении
        return exact['Цена со скидкой, руб.'].sum() / exact['Площадь, кв.м'].sum()
    else:
        # Ищем квартиры с площадью в интервале ±1 кв.м
        approx = match[(match['Площадь, кв.м'] >= area - 2) & (match['Площадь, кв.м'] <= area + 2)]

        if not approx.empty:
            # Средневзвешенная цена по площади
            return approx['Цена со скидкой, руб.'].sum() / approx['Площадь, кв.м'].sum()
        else:
            return np.nan 
```

```python
pipin_april['Цена'] = pipin_april.apply(lambda row: get_weighted_price(row, df_april), axis=1)
```
