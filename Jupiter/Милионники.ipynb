```python
def to_quarter_format(value):
    # Если пустое значение — оставляем NaN
    if pd.isna(value) or str(value).strip() == '':
        return pd.NA

    s = str(value).strip()

    # Если уже в виде "N квартал YYYY года" — оставляем как есть
    if 'квартал' in s:
        return s

    try:
        # Пробуем преобразовать в datetime
        dt = pd.to_datetime(s)
        quarter = (dt.month - 1) // 3 + 1
        return f"{quarter} квартал {dt.year} года"
    except:
        # Если не дата и не текстовый квартал — оставляем как есть
        return s

# Применяем к колонке
df['Срок сдачи'] = df['Срок сдачи'].apply(to_quarter_format)
```
