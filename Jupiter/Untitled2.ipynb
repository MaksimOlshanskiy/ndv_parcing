```python
import pandas as pd
import numpy as np
```

```python
name = '09-10.2024_МО'
df1 = pd.read_excel(f"2024/{name}.xlsb", sheet_name="массив", engine='pyxlsb')
```

```python
df1
```

```python
df1['Дата обновления'] = pd.to_datetime('1899-12-30') + pd.to_timedelta(df1['Дата обновления'], unit='D')
```

```python
df1 = df1[(df1['Дата обновления'].dt.month == 10) & (df1['Дата обновления'].dt.year == 2024)] 
```

```python
df1.to_excel(f"2024/{name}.xlsx")
```
