```python
baza = pd.read_excel("04-05.2025_РЫНОК_типология_fixed.xlsx")
```

```python
baza = baza.drop_duplicates()
```

```python
merged_baza = pd.read_excel("merged_df.xlsx")
```

```python
baza = baza.merge(
    merged_baza[['Название проекта', 'matched_id']],
    on='Название проекта',
    how='left'
)
```

```python
baza
```

```python
unique_id_counts = df.groupby('project_clean')['id'].nunique()
```

```python
projects_with_one_id = unique_id_counts[
    (unique_id_counts == 1) &    
    (~unique_id_counts.index.str.contains(r'\d', na=False))
].index
```

```python
projects_with_one_id
```

```python
filtered_df = df[df['project_clean'].isin(projects_with_one_id)]
```

```python
filtered_df
```

```python
def normalize_project(name):
    return str(name).lower().replace('"', '').replace("'", '').replace(".", ' ').replace(",", '').replace("-", ' ').replace("жилой комплекс", '').replace("жк", '').strip()
```

```python
df['project_clean'] = df['Название проекта'].apply(normalize_project)
baza['project_clean'] = baza['Название проекта'].apply(normalize_project)
```

```python
# Создадим новую колонку для результатов
baza['matched_id'] = np.nan

# Проход по строкам baza
for i, baza_row in baza.iterrows():
    for _, filt_row in filtered_df.iterrows():
        if baza_row['project_clean'] in filt_row['project_clean']:
            baza.at[i, 'matched_id'] = filt_row['id']
            break  # остановиться на первом совпадении
```
