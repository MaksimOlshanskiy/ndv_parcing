```python
def get_weighted_price(row, df_may):
    project = str(row['ЖК рус']).strip().lower()
    area = row['Площадь']
    developer = str(row['Застройщик ЖК']).strip().lower()

    # Список проектов, для которых нужно учитывать девелопера
    special_projects = {
        'заречье',
        'парк люберцы',
        'новоград павлино',
        'павелецкая сити',
        'одинград. квартал семейный'
    }

    if project in special_projects:
        match = df_may[
            (df_may['Название проекта'] == project) &
            (df_may['Девелопер'] == developer)
        ]
    else:
        match = df_may[
            df_may['Название проекта'] == project
        ]

    # Поиск точного совпадения по площади
    exact = match[np.isclose(match['Площадь, кв.м'], area, atol=0.01)]

    if not exact.empty:
        return exact['Цена со скидкой, руб.'].sum() / exact['Площадь, кв.м'].sum()
    
    approx = match[
        (match['Площадь, кв.м'] >= area - 2) &
        (match['Площадь, кв.м'] <= area + 2)
    ]

    if not approx.empty:
        return approx['Цена со скидкой, руб.'].sum() / approx['Площадь, кв.м'].sum()

    return np.nan
```

```python
pipin_may['СВЦ'] = pipin_may.apply(lambda row: get_weighted_price(row, df_may), axis=1)
```
